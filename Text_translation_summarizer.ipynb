```python
from transformers.utils import logging
logging.set_verbosity_error()
```

```python
from transformers import pipeline
import torch
```

```python
translator_ppl = pipeline(task="translation",
                      model="facebook/nllb-200-distilled-600M",
                      torch_dtype=torch.bfloat16)
```

```python
text = """\
In a bright forest, Timmy the turtle found a magic berry.
One bite and he could talk to animals! He chatted with a squirrel,
danced with a butterfly, and laughed with a bear. As night fell,
 he wished to be home. "Magic berry, take me back!" he said. And poof!
 Timmy was back by his stream, ready to dream of his next big adventure.
 """
```

```python
text_translated = translator_ppl(text,
                             src_lang="eng_Latn",
                             tgt_lang="tam_Taml")
```

```python
text_translated
```

```python
summarizer = pipeline(task="summarization",
                      model="facebook/bart-large-cnn",
                      torch_dtype=torch.bfloat16)
```

```python
text = """
The tower is 324 metres (1,063 ft) tall, about the same height as an 81-storey building,
and the tallest structure in Paris. Its base is square, measuring 125 metres (410 ft)
on each side. During its construction, the Eiffel Tower surpassed the Washington Monument
to become the tallest man-made structure in the world, a title it held for 41 years until
the Chrysler Building in New York City was finished in 1930. It was the first structure
to reach a height of 300 metres. Due to the addition of a broadcasting aerial at the top
of the tower in 1957, it is now taller than the Chrysler Building by 5.2 metres (17 ft).
Excluding transmitters, the Eiffel Tower is the second tallest free-standing structure
in France after the Millau Viaduct.
"""
```

```python
summary = summarizer(text,
                     min_length=50,
                     max_length=100)
```

```python
import gradio as gr
import os

def translator_func(text):
    text_translated = translator_ppl(text,
                             src_lang="eng_Latn",
                             tgt_lang="tam_Taml")
    return text_translated[0]['translation_text']

def summarizer_func(text):
    summary = summarizer(text,
                     min_length=50,
                     max_length=100)
    return summary[0]['summary_text']

translate = gr.Interface(
    fn=translator_func,
    inputs=["text"],
    outputs=["text"],
)

summarize = gr.Interface(
    fn=summarizer_func,
    inputs=["text"],
    outputs=["text"],
)

demo = gr.Blocks()

with demo:
    gr.TabbedInterface(
      [translate,
       summarize],
      ["Translate English to Tamil",
       "Summarize text"]
)

demo.launch(share=True,
            debug=True)
```

```python
demo.close()
```
